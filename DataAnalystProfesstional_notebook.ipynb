```python
import pandas as pd
data = pd.read_csv('product_sales.csv')
```

```python
data.info()
```

```python
data.describe()
```

```python
# Display unique values for each column
for column in data.columns:
    unique_values = data[column].unique()
    print(f"Unique values in column '{column}':")
    print(unique_values[:])  
    print(f"Total unique values: {len(unique_values)}\n")
```

```python
data[data['revenue'].isnull()]
```

```python
data['revenue'] = data['revenue'].fillna(93.935)
```

```python
# Replace all values in 'years_as_customer' greater than 40 with 40
data.loc[data['years_as_customer'] > 40, 'years_as_customer'] = 40
```

```python
 import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for 'revenue'
plt.figure(figsize=(10, 6))
sns.histplot(data['revenue'], kde=True, bins=30, color='blue')
plt.title('Distribution of Revenue')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.show()

# Histogram for 'nb_site_visits'
plt.figure(figsize=(10, 6))
sns.histplot(data['nb_site_visits'], kde=True, bins=30, color='green')
plt.title('Distribution of Number of Site Visits')
plt.xlabel('Number of Site Visits')
plt.ylabel('Frequency')
plt.show()
```

```python

# Standardize the 'sales_method' column
data['sales_method'] = data['sales_method'].str.lower().str.strip()

# Replace 'em + call' and other possible variants
data['sales_method'] = data['sales_method'].replace(['em + call', 'em+call', 'email + call'], 'email+call')

# Check for unique sales methods to confirm standardization
print("Unique sales methods after standardization:", data['sales_method'].unique())

import seaborn as sns
# Group and calculate revenue for each sales method
sales_method_revenue = data.groupby('sales_method')['revenue'].sum().reset_index()

# Group and calculate revenue for each state
state_revenue = data.groupby('state')['revenue'].sum().reset_index()

# Group and calculate revenue by number of site visits
site_visits_revenue = data.groupby('nb_site_visits')['revenue'].sum().reset_index()

# Plot Revenue by Sales Method
plt.figure(figsize=(10, 5))
sns.barplot(data=sales_method_revenue, x='sales_method', y='revenue', palette='Blues_d')
plt.title('Revenue by Sales Method')
plt.xlabel('Sales Method')
plt.ylabel('Revenue')
plt.show()

# # Plot Revenue by State
# plt.figure(figsize=(10, 5))
# sns.barplot(data=state_revenue, x='state', y='revenue', palette='Greens_d')
# plt.title('Revenue by State')
# plt.xlabel('State')
# plt.ylabel('Revenue')
# plt.show()

# Plot Revenue by Number of Site Visits
plt.figure(figsize=(10, 5))
sns.lineplot(data=site_visits_revenue, x='nb_site_visits', y='revenue', marker='o', color='orange')
plt.title('Revenue by Number of Site Visits')
plt.xlabel('Number of Site Visits')
plt.ylabel('Revenue')
plt.show()

# Identify Best Sales Method
best_sales_method = sales_method_revenue.loc[sales_method_revenue['revenue'].idxmax()]
print(f"The best sales method is {best_sales_method['sales_method']} with revenue of {best_sales_method['revenue']}.")
```

```python
# Get top 3 states by revenue
top_3_states = state_revenue.sort_values(by='revenue', ascending=False).head(3)

# Plot Revenue for Top 3 States
plt.figure(figsize=(8, 5))
sns.barplot(data=top_3_states, x='state', y='revenue', palette='Greens_d')
plt.title('Top 3 States by Revenue')
plt.xlabel('State')
plt.ylabel('Revenue')
plt.show()
```

```python
# Group and calculate mean and count of revenue for each sales method
sales_method_revenue_stats = data.groupby('sales_method')['revenue'].agg(['sum','mean', 'count']).reset_index()

# Group and calculate mean and count of revenue for each state
state_revenue_stats = data.groupby('state')['revenue'].agg(['sum','mean', 'count']).reset_index()

# Group and calculate mean and count of revenue by number of site visits
site_visits_revenue_stats = data.groupby('nb_site_visits')['revenue'].agg(['sum','mean', 'count']).reset_index()

# Display results
print("Sales Method Revenue Statistics:")
print(sales_method_revenue_stats)

print("\nState Revenue Statistics:")
print(state_revenue_stats)

print("\nSite Visits Revenue Statistics:")
print(site_visits_revenue_stats)
```

```python

# Group and calculate revenue for each sales method
sales_method_revenue_mean = data.groupby('sales_method')['revenue'].mean().reset_index()

# Group and calculate revenue for each state
state_revenue_mean = data.groupby('state')['revenue'].mean().reset_index()

# Group and calculate revenue by number of site visits
site_visits_revenue_mean = data.groupby('nb_site_visits')['revenue'].mean().reset_index()

# Plot Revenue by Sales Method
plt.figure(figsize=(10, 5))
sns.barplot(data=sales_method_revenue_mean, x='sales_method', y='revenue', palette='Blues_d')
plt.title('Average Revenue by Sales Method')
plt.xlabel('Sales Method')
plt.ylabel('Revenue')
plt.show()


# Plot Revenue by Number of Site Visits
plt.figure(figsize=(10, 5))
sns.lineplot(data=site_visits_revenue_mean, x='nb_site_visits', y='revenue', marker='o', color='orange')
plt.title('Average Revenue by Number of Site Visits')
plt.xlabel('Number of Site Visits')
plt.ylabel('Revenue')
plt.show()

# Identify Best Sales Method
best_sales_method = sales_method_revenue_mean.loc[sales_method_revenue_mean['revenue'].idxmax()]
print(f"The best sales method is {best_sales_method['sales_method']} with revenue of {best_sales_method['revenue']}.")
```

```python
# Calculate total revenue per customer
data['total_revenue_per_customer'] = data['revenue'] / data['nb_sold']

# Group by customer_id and calculate the average revenue per product
avg_revenue_per_customer = data.groupby('customer_id')['total_revenue_per_customer'].mean().reset_index()

# Check the first few rows of the new data
avg_revenue_per_customer.head()
import seaborn as sns
import matplotlib.pyplot as plt

# Plot the distribution of average revenue per customer per product
plt.figure(figsize=(10, 6))
sns.histplot(avg_revenue_per_customer['total_revenue_per_customer'], kde=True, bins=30, color='purple')
plt.title('Distribution of Average Revenue per Customer per Product')
plt.xlabel('Average Revenue per Customer per Product')
plt.ylabel('Frequency')
plt.show()

# Optionally, you can use a boxplot to see the spread and outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=avg_revenue_per_customer, x='total_revenue_per_customer', color='lightblue')
plt.title('Box Plot of Average Revenue per Customer per Product')
plt.xlabel('Average Revenue per Customer per Product')
plt.show()
```
